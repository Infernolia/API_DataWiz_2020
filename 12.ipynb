```python
# import yfinance library for financial information of companies at the stock market

import yfinance as yf

# import other necessary libraries

import matplotlib.pyplot as plt
```

```python
# import company data into variables

msft = yf.Ticker("MSFT")
amzn = yf.Ticker("AMZN")
goog = yf.Ticker("GOOG")
fb = yf.Ticker("FB")
aapl = yf.Ticker("AAPL")

# take data of stock prices into respective company variables

microsoft = msft.history(start="2014-10-01", end="2018-10-01")
amazon = amzn.history(start="2014-10-01", end="2018-10-01")
facebook = fb.history(start="2014-10-01", end="2018-10-01")
google = goog.history(start="2014-10-01", end="2018-10-01")
apple = aapl.history(start="2014-10-01", end="2018-10-01")
```

```python
# resample the data into groups of 1 month

aapl_df = apple.resample('1M').mean()
msft_df = microsoft.resample('1M').mean()
goog_df = google.resample('1M').mean()
fb_df = facebook.resample('1M').mean()
amzn_df = amazon.resample('1M').mean()

# plot a graph comparing the highest stock value of companies over the years

# set graph style
plt.style.use('seaborn')

# make subplots

fig, ((ax1,ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize=(15,10), sharey=True, sharex=True)

# plot the values in respective subplots

ax1.plot(fb_df["Open"], label = 'Facebook')
ax1.plot(msft_df["Open"], label = 'Microsoft')
ax1.plot(aapl_df["Open"], label = 'Apple')
ax1.plot(goog_df["Open"], label = 'Google')
ax1.plot(amzn_df["Open"], label = 'Amazon')

ax2.plot(fb_df["Close"], label = 'Facebook')
ax2.plot(msft_df["Close"], label = 'Microsoft')
ax2.plot(aapl_df["Close"], label = 'Apple')
ax2.plot(goog_df["Close"], label = 'Google')
ax2.plot(amzn_df["Close"], label = 'Amazon')

ax3.plot(fb_df["High"], label = 'Facebook')
ax3.plot(msft_df["High"], label = 'Microsoft')
ax3.plot(aapl_df["High"], label = 'Apple')
ax3.plot(goog_df["High"], label = 'Google')
ax3.plot(amzn_df["High"], label = 'Amazon')

ax4.plot(fb_df["Low"], label = 'Facebook')
ax4.plot(msft_df["Low"], label = 'Microsoft')
ax4.plot(aapl_df["Low"], label = 'Apple')
ax4.plot(goog_df["Low"], label = 'Google')
ax4.plot(amzn_df["Low"], label = 'Amazon')

# label the figure, subplots and set grid

fig.suptitle('Stock Value comparisons', fontsize=16 ,x=0.5, y=1.1)

ax1.legend()
ax1.set_title('Comparison of opening stock values of companies over the years.')
ax1.set_ylabel('Stock Value')

ax2.legend()
ax2.set_title('Comparison of closing stock values of companies over the years.')

ax3.legend()
ax3.set_title('Comparison of highest stock values of companies over the years.')
ax3.set_xlabel('Years')
ax3.set_ylabel('Stock Value')

ax4.legend()
ax4.set_title('Comparison of lowest stock values of companies over the years.')
ax4.set_xlabel('Years')

plt.grid(True)
fig.tight_layout()

# display graph

plt.show()

# CONCLUSION: Amazon's stock had the best performance over the course of 4 years, while Microsoft had the poorest performance.
```

```python
# comparing the maximum 'high' stock value of all companies

# take max values in variables

msft_max = msft_df['High'].max()
amzn_max = amzn_df['High'].max()
fb_max = fb_df['High'].max()
aapl_max = aapl_df['High'].max()
goog_max = goog_df['High'].max()

# set x, y axes and width of bar

company = ['Microsoft', 'Amazon', 'Facebook', 'Apple', 'Google']
value = [msft_max, amzn_max, fb_max, aapl_max, goog_max]
width = 0.35

# plot bar graph with above parameters

plt.bar(company, value, width)

# label plot

plt.title('Maximum stock value', fontsize=16)
plt.ylabel('Stock Value')

# show graph

plt.tight_layout()
plt.show()

# CONCLUSION: Amazon has had the maximum 'high' stock value in the given period of time
```
