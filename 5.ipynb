```python
import pandas as pd
```

```python
import pandas_datareader.data as web
```

```python
import datetime as dt
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

## From 2015 to 2020 

```python
start = dt.datetime(2015,1,1)
end = dt.datetime(2020,8,14)
```

## YAHOO API'S TESLA STOCKS

```python
df = web.DataReader('TSLA','yahoo',start,end)
df.to_csv("TSLA.csv")
df.describe()
```

```python
df.tail()
```

```python
df.head()
```

```python
f, ax = plt.subplots(figsize=(10, 6.5))
sns.despine(f, left=False, bottom=False)
sns.scatterplot(x = 'Close', y ='Open', hue = 'High', sizes=(1, 8), data = df)
plt.show()
```

```python
df['Adj Close'].plot()
plt.xlabel("Years")
plt.ylabel("Adj Close")
plt.show()
```

```python
df['Close'].plot()
plt.xlabel("Years")
plt.ylabel("Close")
plt.show()
```

```python
highh = df['High']
highh1 = highh/7
highh1.plot()
plt.show()
```

```python
g = df['Volume']
Vol = g/10000
Vol.plot()
```

## YAHOO API'S FACEBOOK STOCKS

```python
df1 = web.DataReader('fb','yahoo',start,end,)
df1.tail()
df1.describe()
```

```python
df1['Close'].plot()
```

## DOWNLOADING FACEBOOK STOCKS IN CSV FILE.

```python
df1.to_csv("facebook.csv")
```

```python
s = df1['Volume']
volume1 = s/10000
volume1
```

```python
f, ax = plt.subplots(figsize=(10, 6.5))
sns.despine(f, left=False, bottom=False)
sns.scatterplot(x = 'Close', y ='Open', hue = 'Low', sizes=(1, 8), data = df1)
plt.show()
```

## COMPARING VOLUME OF FACEBOOK AND TESLA OVER THE YEARS

```python
volume1.plot()
Vol.plot()
plt.legend(['Facebook high','Tesla high'])
plt.xlabel('years')
plt.ylabel('Volume/10000')
plt.rcParams["figure.figsize"] = (10,10)
plt.show()
```

## COMPARING HIGH STOCK OF FACEBOOK AND TESLA OVER THE YEARS

```python
df1['High'].plot()
df['High'].plot()
plt.xlabel("years")
plt.ylabel("High")
plt.legend(['Facebook high','Tesla high'])
plt.rcParams["figure.figsize"] = (20,20)
plt.show()
```

## YAHOO API'S NETFLIX STOCKS

```python
df2 = web.DataReader('NFLX','yahoo',start,end)
df2.tail()
df2.to_csv("Netflix.csv")
```

```python
f,ax = plt.subplots(figsize = (10,7.5))
sns.despine(f, left = True, bottom = True)
sns.scatterplot(x = 'Close', y = 'Open' , hue = 'Low', sizes=(1,8), data = df2)
```
